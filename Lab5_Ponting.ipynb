```python
#2 What is the average departure delay (`dep_delay`) of all flights in this data?

flights.agg({'dep_delay':['mean']})
```

```python
#3  What is the average departure delay by carrier, for flights in this data?
flights.groupby('carrier').agg({'dep_delay':['mean']})
```

```python
#4 If you followed the groupby-agg workflow we covered in lecture, you passed your summary function within a list.Try removing the brackets and rerunning #3.,
# putting in a list allows you to get multiple aggregations at once 

flights.groupby('carrier').agg({'dep_delay':'mean'})
```

```python
#5 Working from your code for #3, calculate the minimum, mean, median, and maximum departure delay for each carrier.,
#Don't do this in 4 separate lines -- you can achieve this result using a single invocation of `groupby` and a single invocation of `agg`

flights.groupby('carrier').agg({'dep_delay':['mean','min','median','max']})
```

```python
#6 Perhaps, after doing some research, you discover that the most meaningful statistic for departure delay is the *average delay*, but the most meaningful way to measure 
# air time is the *median air time. Build a groupby-agg invocation that summarizes – at the carrier level – the average departure delay and the median air time.

flights.groupby('carrier').agg({'dep_delay':['mean'],'air_time':['median']})
```

```python
#7 What is the single most common route in this data? That is to say, what combination of values for `origin` and `destination` occurs most frequently? 
# Hint: to order the data yourself, you may want to experiment with the `sort_values` method of Pandas Series.

flights.groupby(['origin','dest']).size().reset_index().rename(columns={0:'count'}).sort_values(by='count',ascending=False)
```
