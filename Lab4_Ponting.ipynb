```python
#4 Just using the flights data, is it possible to determine what percentage of the flights were completed on Boeing planes? Why not?
# No, there is not a column with the manufacturere in the flights data set. 
```

```python
#5 In order to combine this data with the planes data (where the manufacturer column is), we need to do a join.
#Inner join the planes data to the flights data, storing the result in a variable called joined_data.
joined_data = pd.merge(flights, planes,on='tailnum')
joined_data
```

```python
#7 What would happen if you did a left join instead of an inner join? Try it, storing the result in a variable called left_joined_data. 
#Check how many rows are in this new DataFrame -- is it different to the number of rows in joined_data (from #5)?

left_joined_data = pd.merge(flights, planes,on='tailnum',how='outer')
left_joined_data.shape

# Left outter join grabs all of the rows while inner join only grabs the rows that are in common. 
```

```python
#8 Using the original joined data, calculate what percentage of flights were completed on Boeing planes.
(joined_data.loc[joined_data['manufacturer'] == 'BOEING'].count() / joined_data['tailnum'].count()) * 100
```
