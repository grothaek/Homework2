```python
#3 Experiment with the .dropna method -- consult its documentation using the ?
#Use it to drop destinations in your result from #2 where at least one value is NaN. Store this "cleaned" DataFrame in a variable called "clean_df".
#destinations may be either rows or columns depending on how you did #2

clean_df = pivoted_df.dropna(axis=1,how='any',thresh=None,subset=None, inplace=False)
clean_df
```
